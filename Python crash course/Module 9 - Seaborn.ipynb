# Distribution plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde') # kind hex, reg, kde
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') 
# Pair plots for every combination of every numerical variable
# Hue - сравнить данные с помощью категориальных переменных
# Palette -  цветовая схема графиков
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorial Plots

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
# Estimator - по какому статистическому показателю сравниваются переменные
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
# Split - сформировать сравниваемы категории в hue в едином графике
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```
