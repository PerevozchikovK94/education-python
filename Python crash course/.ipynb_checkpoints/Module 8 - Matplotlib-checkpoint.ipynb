```python
import matplotlib.pyplot as plt
```

```python
# Turn on graphics in Jupyter notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# Functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Multi graphic
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes1.set_title('Larger plot')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout()
```

```python
for current_ax in axes:
    current_ax.plot(x,y)
```

# Figure size and DPI

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8,2), dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('mypicture.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend(loc=0)
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00', lw=3, alpha=0.5, ls='--', marker='o', markersize=10, 
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green') # RGB Hex code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00', lw=2, ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
